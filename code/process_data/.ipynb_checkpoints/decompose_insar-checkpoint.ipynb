```python
def get_dir_size(path):
    total = 0
    largest_file = None
    largest_size = 0
    for dirpath, dirnames, filenames in os.walk(path):
        for f in filenames:
            fp = os.path.join(dirpath, f)
            if os.path.isfile(fp):
                s = os.path.getsize(fp)
                total += s
                if s > largest_size:
                    largest_size = s
                    largest_file = fp
    return total, largest_file, largest_size

def human_readable(size):
    for unit in ['B', 'K', 'M', 'G', 'T']:
        if size < 1024:
            return f"{size:.1f}{unit}"
        size /= 1024

sizes = {d: get_dir_size(data_storage / d) for d in completed_aquifers}
grand_total = sum(v[0] for v in sizes.values())

for d, (s, lf, ls) in sorted(sizes.items(), key=lambda x: x[1][0], reverse=True):
    print(f"{human_readable(s)}\t{d}\t(largest: {human_readable(ls)}  {lf})")

print(f"{human_readable(grand_total)}\ttotal")

# Overall largest file across all aquifers
biggest_dir = max(sizes, key=lambda d: sizes[d][2])
_, global_largest_file, global_largest_size = sizes[biggest_dir]
print(f"\nLargest single file: {human_readable(global_largest_size)}  {global_largest_file}")
```

```python
test_aquifer = completed_aquifers[30]
# test_aquifer = 'N034'
print(test_aquifer)

aquifer_storage = data_storage / str(test_aquifer)

velo_files = list(aquifer_storage.glob('*/*/velocity.h5'))
ts_files = list(aquifer_storage.glob('*/*/timeseries.h5'))

atr_first = readfile.read_attribute(str(velo_files[0]))
raster_epsg = atr_first.get('EPSG', 'N/A')
aquifer = gdf[gdf.CODE_2021 == aquifer_storage.stem].copy().to_crs(raster_epsg)

# Get US-Mexico border from Natural Earth
shpfile = shpreader.natural_earth(resolution='10m', category='cultural',
                                   name='admin_0_boundary_lines_land')
borders_gdf = gpd.read_file(shpfile).to_crs(raster_epsg)

valid_files = []
for vf in velo_files:
    velo, atr = readfile.read(str(vf))
    if not np.all(np.isnan(velo)):
        valid_files.append(vf)

n = len(valid_files)
fig, axes = plt.subplots(1, n, figsize=(7 * n, 6))
if n == 1:
    axes = [axes]

for i, velo_file in enumerate(valid_files):
    velo, atr_velo = readfile.read(str(velo_file))
    velo_cm_yr = velo * 100
    frame_epsg = int(atr_velo.get('EPSG'))
    vlim = np.max(np.abs(velo_cm_yr))  # dynamic per-frame limit

    # Reproject overlays to match this frame's CRS
    aquifer_proj = aquifer.to_crs(epsg=frame_epsg)
    borders_proj = borders_gdf.to_crs(epsg=frame_epsg)

    # Check if raster actually overlaps the aquifer
    x0 = float(atr_velo['X_FIRST'])
    xs = float(atr_velo['X_STEP'])
    y0 = float(atr_velo['Y_FIRST'])
    ys = float(atr_velo['Y_STEP'])
    length, width = velo.shape
    raster_box = box(x0, y0 + ys * length, x0 + xs * width, y0)
    if not aquifer_proj.union_all().intersects(raster_box):
        axes[i].set_title('No overlap — skipped')
        continue

    extent = [x0, x0 + xs * width, y0 + ys * length, y0]
    im = axes[i].imshow(velo_cm_yr, cmap='RdBu', extent=extent,
                         interpolation='nearest', vmin=-vlim, vmax=vlim)

    aquifer_proj.boundary.plot(ax=axes[i], edgecolor='black', linewidth=1.5, zorder=10)

    xlim, ylim = axes[i].get_xlim(), axes[i].get_ylim()
    borders_proj.clip_by_rect(xlim[0], ylim[0], xlim[1], ylim[1]).plot(
        ax=axes[i], edgecolor='red', linewidth=2, zorder=11
    )

    axes[i].set_xlabel('Easting (m)')
    axes[i].set_ylabel('Northing (m)')
    orbit_dir = atr_velo.get('ORBIT_DIRECTION', 'Unknown')
    axes[i].set_title(f'{orbit_dir} LOS Velocity')
    plt.colorbar(im, ax=axes[i], label='cm/yr', shrink=0.7)

plt.tight_layout()
plt.show()
```

# Decompose from LOS displacement to East-West and Up-Down displacement using acending and descending tracks via MintPy

```python
asc_list = []
desc_list = []

for velo_file in list(aquifer_storage.glob('*/*/velocity.h5')):
    orb_dir = xr.open_dataset(velo_file, engine='h5netcdf', phony_dims = 'sort').attrs['ORBIT_DIRECTION']
    if orb_dir == 'Ascending':
        asc_list.append(velo_file)
    else:
        desc_list.append(velo_file)

asc_attrs = [readfile.read_attribute(str(f)) for f in asc_list]
desc_attrs = [readfile.read_attribute(str(f)) for f in desc_list]

try:
    for i, asc_file in enumerate(asc_list):
        for j, desc_file in enumerate(desc_list):
            bbox = get_overlap_lalo([asc_attrs[i], desc_attrs[j]])
            S, N, W, E = bbox

            lat_step = float(asc_attrs[i]['Y_STEP'])
            lon_step = float(asc_attrs[i]['X_STEP'])
            length = int(round((S - N) / lat_step))
            width = int(round((E - W) / lon_step))

            if (length > 0 and width > 0):
                inps = Namespace(
                    file=[asc_file, desc_file],
                    geom_file=[asc_file.parent / 'geometryGeo.h5', desc_file.parent / 'geometryGeo.h5'],
                    ds_name=None,
                    horz_az_angle=-90.0,
                    outfile=[str(aquifer_storage / f'{asc_file.parts[-3]}_{desc_file.parts[-3]}_horizontalDefo.h5'), str(aquifer_storage/ f'{asc_file.parts[-3]}_{desc_file.parts[-3]}_verticalDefo.h5')],
                    one_outfile = None          
                )

                run_asc_desc2horz_vert(inps)
    # decomposed_aquifers.append(aquifer)
        
except Exception as e:
    print(f"Decomposing aquifer {aquifer} failes due to: {e}")
```

```python
horz_defo_files = list(aquifer_storage.glob('*horizontalDefo*'))
vert_defo_files = list(aquifer_storage.glob('*verticalDefo*'))

# Before the loop, find global color limits
all_horz_vals = []
all_vert_vals = []
for hf, vf in zip(horz_defo_files, vert_defo_files):
    h = readfile.read(str(hf))[0] * 100
    v = readfile.read(str(vf))[0] * 100
    all_horz_vals.extend([np.nanmin(h), np.nanmax(h)])
    all_vert_vals.extend([np.nanmin(v), np.nanmax(v)])

horz_vlim = max(abs(min(all_horz_vals)), abs(max(all_horz_vals)))
vert_vlim = max(abs(min(all_vert_vals)), abs(max(all_vert_vals)))
```

```python
def get_pair_name(f):
    return f.stem.replace('_horizontalDefo', '').replace('_verticalDefo', '')

horz_dict = {get_pair_name(f): f for f in horz_defo_files}
vert_dict = {get_pair_name(f): f for f in vert_defo_files}

matched_pairs = []
for pair_name in horz_dict:
    if pair_name in vert_dict:
        matched_pairs.append((pair_name, horz_dict[pair_name], vert_dict[pair_name]))
        print(f"Matched: {pair_name}")
```

```python
# TODO: Maybe try and plot all of them in one go? Maybe not feasible, but would make a super cool plot

# Get the aquifer polygon
atr_first = readfile.read_attribute(str(horz_defo_files[0]))
raster_epsg = atr_first.get('EPSG', 'N/A')
aquifer = gdf[gdf.CODE_2021 == aquifer_storage.stem].copy().to_crs(raster_epsg)

# Get US-Mexico border from Natural Earth
shpfile = shpreader.natural_earth(resolution='10m', category='cultural',
                                   name='admin_0_boundary_lines_land')
borders_gdf = gpd.read_file(shpfile).to_crs(raster_epsg)

# Compute global color limits
all_horz, all_vert = [], []
for hf, vf in zip(horz_defo_files, vert_defo_files):
    h = readfile.read(str(hf))[0] * 100
    v = readfile.read(str(vf))[0] * 100
    if not np.all(np.isnan(h)):
        all_horz.extend([np.nanmin(h), np.nanmax(h)])
    if not np.all(np.isnan(v)):
        all_vert.extend([np.nanmin(v), np.nanmax(v)])
horz_vlim = max(abs(min(all_horz)), abs(max(all_horz)))
vert_vlim = max(abs(min(all_vert)), abs(max(all_vert)))

fig, axes = plt.subplots(1, 2, figsize=(14, 6))

for _, horz_file, vert_file in matched_pairs:
    horz, atr_h = readfile.read(str(horz_file))
    vert, atr_v = readfile.read(str(vert_file))

    horz_cm = horz * 100
    vert_cm = vert * 100

    if np.all(np.isnan(horz_cm)) and np.all(np.isnan(vert_cm)):
        continue

    x0 = float(atr_h['X_FIRST'])
    xs = float(atr_h['X_STEP'])
    y0 = float(atr_h['Y_FIRST'])
    ys = float(atr_h['Y_STEP'])
    length, width = horz.shape
    extent = [x0, x0 + xs * width, y0 + ys * length, y0]

    im0 = axes[0].imshow(horz_cm, cmap='RdBu', extent=extent,
                          interpolation='nearest', vmin=-horz_vlim, vmax=horz_vlim)
    im1 = axes[1].imshow(vert_cm, cmap='RdBu_r', extent=extent,
                          interpolation='nearest', vmin=-vert_vlim, vmax=vert_vlim)

# Overlay aquifer boundary
aquifer.boundary.plot(ax=axes[0], edgecolor='black', linewidth=1.5, zorder=10)
aquifer.boundary.plot(ax=axes[1], edgecolor='black', linewidth=1.5, zorder=10)

# Overlay country borders — clip to plot extent for cleaner rendering
for ax in axes:
    xlim, ylim = ax.get_xlim(), ax.get_ylim()
    borders_gdf.clip_by_rect(xlim[0], ylim[0], xlim[1], ylim[1]).plot(
        ax=ax, edgecolor='red', linewidth=2, zorder=11
    )
    ax.set_xlabel('Easting (m)')
    ax.set_ylabel('Northing (m)')

axes[0].set_title('Horizontal (East-West) Velocity')
axes[1].set_title('Vertical (Up-Down) Velocity')

plt.colorbar(im0, ax=axes[0], label='cm/yr', shrink=0.7)
plt.colorbar(im1, ax=axes[1], label='cm/yr', shrink=0.7)

plt.tight_layout()
plt.show()
```
