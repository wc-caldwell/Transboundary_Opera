```python
# retries are needed as it is quite common that the connect be interrupted or broken
MAX_RETRIES = 5
RETRY_DELAY = 10  # seconds, will increase exponentially

SEARCH_START = datetime(2016, 1, 1)     
SEARCH_END = datetime(2026, 1, 1)

data_storage = Path('/home/clayc/Documents/US_Mex_InSAR/OPERA_data/')

gdf = gpd.read_file(Path(os.getcwd()).parent.parent / "raw_data/TBA_full.shp")
len(gdf)
```

```python
frame_ids = dt.get_unique_frame_ids(gdf, track_per_row=True, max_workers=8)
len(frame_ids)
```

## Step 1. Download the OPERA DISP NetCDF files
- Starting to reach 4 tb capacity for the aquifers. Going to give up and process them individually and then try and remove the no longer needed data.

```python
# aquifers left over to process as of 09 MAR 2026
# ['N087', 'N026', 'N078', 'N074', 'N071']

list(set(list(gdf.CODE_2021.values)) - set(os.listdir(data_storage)))
```

```python
for idx, data in gdf.iterrows():
    print(f"Processing aquifer {data.CODE_2021}...", flush=True)
    bbox = data.geometry.bounds
    frame_ids = data.frame_ids
    print(f"  Fetching frame geometries for {len(frame_ids)} frames...", flush=True)
    geom_frames = dt.get_frame_geometries(
            frame_ids, 
            gdf_bounds=data.geometry.bounds
        )
    print(f"  Got geometries, starting downloads...", flush=True)
    
    for frame in frame_ids:
        out_path = data_storage / data.CODE_2021 / str(frame)
        subset_dir = out_path / 'subset-ncs'
        mintpy_dir = out_path / 'mintpy'
        done_marker = out_path / '.download_complete'
        
        # Skip if already processed by mintpy
        if mintpy_dir.exists():
            print(f'  Frame {frame} for aquifer {data.CODE_2021} already processed (mintpy exists).', flush=True)
            continue
        
        # Skip if download already completed
        if done_marker.exists():
            print(f'  Data from OPERA frame {frame} for aquifer {data.CODE_2021} already downloaded locally.', flush=True)
            continue
        
        # Clean up any partial downloads from previous interrupted runs
        if subset_dir.exists():
            print(f"  Cleaning up partial download for frame {frame}...", flush=True)
            shutil.rmtree(subset_dir)
        
        os.makedirs(out_path, exist_ok=True)
        # clip the target aquifer by the current opera frame
        clipped = gpd.clip(
            gpd.GeoSeries([data.geometry]),
            geom_frames[geom_frames['frame_id'] == frame]
        )
        # if the clipped aquifer is empty, then the frame does not actually overlap and can be removed and skipped
        if clipped.empty:
            print(f"  No overlap for frame {frame}, skipping...", flush=True)
            shutil.rmtree(out_path)
            continue
        
        clipped_bbox = clipped.geometry.iloc[0].bounds
        print(f"  Downloading frame {frame}...", flush=True)
        for attempt in range(1, MAX_RETRIES + 1):
            try:
                _download.run_download(
                    frame,
                    start_datetime=SEARCH_START,
                    end_datetime=SEARCH_END,
                    num_workers=8,
                    output_dir=out_path / 'subset-ncs',
                    bbox=clipped_bbox
                )
                # Mark as complete only after successful download
                done_marker.touch()
                print(f"  Frame {frame} done.", flush=True)
                break  # success
            except (ConnectionError, OSError) as e:
                if attempt < MAX_RETRIES:
                    wait = RETRY_DELAY * (2 ** (attempt - 1))
                    print(f"  Attempt {attempt} failed for frame {frame}: {e}", flush=True)
                    print(f"  Retrying in {wait}s...", flush=True)
                    time.sleep(wait)
                else:
                    print(f"  All {MAX_RETRIES} attempts failed for frame {frame}. Skipping.", flush=True)
```

```python


import gc

# Conservative defaults for lower peak RAM
CHUNKING = (32, 256, 256)      # (time, y, x)
PROCESS_CHUNKS = (512, 512)    # internal processing tile size

DROPPED_VARS = [
    "phase_similarity",
    "persistent_scatterer_mask",
    "timeseries_inversion_residuals",
    "connected_component_labels",
    "estimated_phase_quality",
    "shp_counts",
]

REFERENCE_METHOD = "high_coherence"   # "high_coherence", "point", or "border"
COHERENCE_THRESHOLD = 0.7
BORDER_PIXELS = 0.3
RELIABILITY_THRESHOLD = 0.9


def _cleanup(*objs):
    for obj in objs:
        try:
            del obj
        except Exception:
            pass
    gc.collect()


for aquifer in sorted(os.listdir(data_storage)):
    print(f"Processing aquifer: {aquifer}")
    print("=" * 100)

    for frame in sorted(os.listdir(data_storage / aquifer)):
        if frame == "3067":
            continue

        try:
            print(f"Processing frame: {frame}")
            print("=" * 100)
            out_path = data_storage / aquifer / str(frame)

            if "mintpy" in os.listdir(out_path):
                print(f"Pre-processing for LOS decomposition for frame {frame} already completed.")
                continue

            nc_files = sorted(out_path.glob("*/*.nc"))
            if not nc_files:
                print(f"No NetCDF files found for frame {frame}; skipping.")
                continue

            _reformat.reformat_stack(
                input_files=nc_files,
                output_name=str(out_path / f"disp-stack-{out_path.stem}.nc"),
                out_chunks=CHUNKING,
                shard_factors=(1, 4, 4),
                drop_vars=DROPPED_VARS,
                apply_solid_earth_corrections=True,
                apply_ionospheric_corrections=True,
                reference_method=REFERENCE_METHOD,
                reference_coherence_threshold=COHERENCE_THRESHOLD,
                process_chunk_size=PROCESS_CHUNKS,
                do_round=True,
            )
            gc.collect()

            frame_nc = out_path / f"disp-stack-{frame}.nc"

            staticDir = out_path / "orbit_data"
            os.makedirs(staticDir, exist_ok=True)
            dispDir = out_path / "disp_data"
            os.makedirs(dispDir, exist_ok=True)
            geomDir = out_path / "geom_data"
            os.makedirs(geomDir, exist_ok=True)

            sys.argv = [
                "notebook",
                "--frameID", str(frame),
                "--startDate", SEARCH_START.strftime("%Y%m%d"),
                "--endDate", SEARCH_END.strftime("%Y%m%d"),
                "--dispDir", str(dispDir),
                "--staticDir", str(staticDir),
                "--geomDir", str(geomDir),
                "--staticOnly",
            ]

            for attempt in range(1, MAX_RETRIES + 1):
                try:
                    run1_inps = run1_download_DISP_S1_Static.createParser()
                    run1_download_DISP_S1_Static.main(run1_inps)
                    break  # important: do not rerun successful download N times
                except (ConnectionError, OSError) as e:
                    if attempt < MAX_RETRIES:
                        wait = RETRY_DELAY * (2 ** (attempt - 1))
                        print(f"Attempt {attempt} failed for frame {frame}: {e}")
                        print(f"Retrying in {wait}s...")
                        time.sleep(wait)
                    else:
                        print(f"All {MAX_RETRIES} attempts failed for frame {frame}. Skipping.")
                        raise

            with rio.open(geomDir / "los_east.tif") as _src:
                losEast = _src.read(1).astype(np.float32, copy=False)
                crs = _src.crs
                transform = _src.transform

            with rio.open(geomDir / "los_north.tif") as _src:
                losNorth = _src.read(1).astype(np.float32, copy=False)

            up = np.sqrt(np.clip(1.0 - losEast**2 - losNorth**2, 0.0, 1.0)).astype(np.float32, copy=False)

            with rio.open(
                geomDir / "los_enu.tif",
                "w",
                driver="GTiff",
                height=losEast.shape[0],
                width=losEast.shape[1],
                count=3,
                dtype="float32",
                crs=crs,
                transform=transform,
            ) as dst:
                dst.write(losEast, 1)
                dst.write(losNorth, 2)
                dst.write(up, 3)

            _cleanup(losEast, losNorth, up)

            # Clip geometry rasters to match DISP frame extent
            with xr.open_dataset(frame_nc, cache=False) as nc_ds:
                for tif in ["layover_shadow_mask.tif", "los_enu.tif", "los_east.tif", "los_north.tif"]:
                    with rioxarray.open_rasterio(geomDir / tif, chunks={"x": 512, "y": 512}) as tiff_ds:
                        clipped = tiff_ds.rio.reproject_match(nc_ds)
                        clipped.rio.to_raster(geomDir / tif)
                    _cleanup(clipped)

            try:
                os.rmdir(dispDir)
            except OSError:
                pass

            mintpy.disp_nc_to_mintpy(
                str(out_path / f"disp-stack-{out_path.stem}.nc"),
                sample_disp_nc=nc_files[0],
                los_enu_path=geomDir / "los_enu.tif",
                dem_path=None,
                layover_shadow_mask_path=geomDir / "layover_shadow_mask.tif",
                outdir=out_path / "mintpy",
                virtual=True,
                reliability_threshold=RELIABILITY_THRESHOLD,
            )
            gc.collect()

            base = str(data_storage / aquifer / frame / "mintpy")

            with h5py.File(f"{base}/geometryGeo.h5", "r") as geom:
                epsg = dict(geom.attrs)["EPSG"]
            transformer = Transformer.from_crs(f"EPSG:{epsg}", "EPSG:4326", always_xy=True)

            with xr.open_dataset(f"{base}/avgSpatialCoh.h5", engine="h5netcdf", cache=False) as coh_ds:
                coh_ds = coh_ds.rename({"phony_dim_0": "y", "phony_dim_1": "x"})
                coherence = coh_ds["avgSpatialCoh"].values
                max_flat_idx = int(np.nanargmax(coherence))
                y_idx, x_idx = np.unravel_index(max_flat_idx, coherence.shape)

                y_coord = float(coh_ds["y"].isel(y=y_idx).values)
                x_coord = float(coh_ds["x"].isel(x=x_idx).values)

            _cleanup(coherence)

            ref_lon, ref_lat = transformer.transform(x_coord, y_coord)
            attrs = {
                "REF_LAT": float(ref_lat),
                "REF_LON": float(ref_lon),
                "REF_Y": int(y_idx),
                "REF_X": int(x_idx),
            }

            for fname in ["velocity.h5", "timeseries.h5", "geometryGeo.h5"]:
                fpath = f"{base}/{fname}"
                with h5py.File(fpath, "r+") as f:
                    f.attrs.update(attrs)
                print(f"Added to {fpath}")
                print(
                    f"  REF_LAT: {attrs['REF_LAT']:.6f}, REF_LON: {attrs['REF_LON']:.6f}, "
                    f"REF_Y: {attrs['REF_Y']}, REF_X: {attrs['REF_X']}"
                )

            shutil.rmtree(data_storage / aquifer / frame / "geom_data", ignore_errors=True)
            shutil.rmtree(data_storage / aquifer / frame / "orbit_data", ignore_errors=True)
            shutil.rmtree(data_storage / aquifer / frame / "subset-ncs", ignore_errors=True)

            _cleanup()

        except Exception as e:
            print(f"Frame {frame} for aquifer {aquifer} failed: {e}")
            _cleanup()
            continue
```

```python
# TODO: try with parallel processing. It took 10 hours to write 2.5 out of 5 files for aquifer N025. That is insane.
# may need to test without corrections, though that would be less accurate deformation estimates

# from concurrent.futures import ProcessPoolExecutor, as_completed

import psutil

CHUNKING = (64, 512, 512)        # (Time, Height, Width). (4, 256, 256) uses ~ 17GB of my ~90GB, can increase for faster writing
# (64, 1024, 1024) or larger

# Dynamically scale PROCESS_CHUNKS to stay within ~60% of available RAM.
# Rough model: each spatial pixel needs ~(est. time steps * 4 bytes) in memory.
_available_gb = psutil.virtual_memory().available / (1024 ** 3)
_budget_bytes = _available_gb * 0.60 * (1024 ** 3)
_est_time_steps = 300          # conservative upper bound for number of acquisitions
_bytes_per_pixel = _est_time_steps * 4   # float32
_safe_side = int((_budget_bytes / _bytes_per_pixel) ** 0.5)
_safe_side = max(256, min(2048, 2 ** int(np.log2(_safe_side))))   # clamp & round to power-of-2
PROCESS_CHUNKS = (_safe_side, _safe_side)
print(f"Available RAM: {_available_gb:.1f} GB  →  PROCESS_CHUNKS set to {PROCESS_CHUNKS}")

DROPPED_VARS = ['phase_similarity', 'persistent_scatterer_mask', 'timeseries_inversion_residuals', 'connected_component_labels', 'estimated_phase_quality', 'shp_counts']


REFERENCE_METHOD = 'high_coherence'     # 'high_coherence' ,'point', or 'border'

# if reference_method is high_coherence, need to set a coherence threshold
# default = 0.7
COHERENCE_THRESHOLD = 0.7

# if reference_method is border, need to set a number of border pixels used 
# default is 0.3
BORDER_PIXELS = 0.3

RELIABILITY_THRESHOLD = 0.9

if aquifer:
    for frame in os.listdir(data_storage / aquifer):
        if frame == '3067':
            continue
        else:
            out_path = data_storage / aquifer / str(frame)
            
            if 'mintpy' in os.listdir(out_path):
                print(f'Pre-processing for LOS decomposition for frame {frame} already completed.')
                continue

            else:

                nc_files = list(out_path.glob('*/*.nc'))

                _reformat.reformat_stack(
                            input_files = nc_files,
                            output_name = str(out_path / f'disp-stack-{out_path.stem}.nc'),
                            out_chunks = CHUNKING,
                            shard_factors = (1, 4, 4),
                            drop_vars = DROPPED_VARS,                                                   # can drop variables if we want, but will keep all for future analysis. Can also keep the short wavelength displacement
                            apply_solid_earth_corrections = True,
                            apply_ionospheric_corrections = True,
                            reference_method = REFERENCE_METHOD,
                            reference_coherence_threshold = COHERENCE_THRESHOLD,
                            process_chunk_size = PROCESS_CHUNKS,
                            do_round = True                               
                        )

                frame_nc = out_path / f'disp-stack-{frame}.nc'

                staticDir = out_path / 'orbit_data'
                os.makedirs(staticDir, exist_ok=True)
                dispDir = out_path / 'disp_data'
                os.makedirs(dispDir, exist_ok=True)
                geomDir = out_path / 'geom_data'
                os.makedirs(geomDir, exist_ok=True)

                sys.argv = [
                    'notebook',  # placeholder for script name
                    '--frameID', str(frame),
                    # '--version', 0.9,
                    '--startDate', SEARCH_START.strftime('%Y%m%d'),
                    '--endDate', SEARCH_END.strftime('%Y%m%d'),
                    '--dispDir', str(dispDir),
                    '--staticDir', str(staticDir),
                    '--geomDir', str(geomDir),
                    '--staticOnly'                             # default is true, uncomment for false
                ]

                for attempt in range(1, MAX_RETRIES + 1):
                    try:
                        run1_inps = run1_download_DISP_S1_Static.createParser()
                        run1_download_DISP_S1_Static.main(run1_inps)

                    except (ConnectionError, OSError) as e:
                        if attempt < MAX_RETRIES:
                            wait = RETRY_DELAY * (2 ** (attempt - 1))
                            print(f"Attempt {attempt} failed for frame {frame}: {e}")
                            print(f"Retrying in {wait}s...")
                            time.sleep(wait)
                        else:
                            print(f"All {MAX_RETRIES} attempts failed for frame {frame}. Skipping.")
                            # optionally clean up partial downloads:
                            # shutil.rmtree(out_path)

                with rio.open(geomDir / 'los_east.tif') as _src:
                    losEast = _src.read(1)
                    crs = _src.crs
                    transform = _src.transform
                with rio.open(geomDir / 'los_north.tif') as _src:
                    losNorth = _src.read(1)
                up = np.sqrt(1 - losEast**2 - losNorth**2)

                with rio.open(geomDir / 'los_enu.tif', 'w',
                         driver='GTiff',
                         height=losEast.shape[0],
                         width=losEast.shape[1],
                         count=3,
                         dtype=losEast.dtype,
                         crs=crs,
                         transform=transform) as dst:
                    dst.write(losEast, 1)
                    dst.write(losNorth, 2)
                    dst.write(up, 3)

                # clipping the geometry rasters to the same extent as the clipped OPERA DISP frames
                with xr.open_dataset(frame_nc) as nc_ds:
                    for tif in ['layover_shadow_mask.tif', 'los_enu.tif', 'los_east.tif', 'los_north.tif']:
                        with rioxarray.open_rasterio(geomDir / tif) as tiff_ds:
                            clipped = tiff_ds.rio.reproject_match(nc_ds)
                        clipped.rio.to_raster(geomDir / tif)

                os.rmdir(dispDir)   # DISP products are already downloaded

                mintpy.disp_nc_to_mintpy(
                        str(out_path / f'disp-stack-{out_path.stem}.nc'),
                        sample_disp_nc = nc_files[0],
                        los_enu_path = geomDir / 'los_enu.tif',
                        dem_path = None,
                        layover_shadow_mask_path = geomDir / 'layover_shadow_mask.tif',
                        outdir = out_path / "mintpy"    ,
                        virtual = True,
                        reliability_threshold = RELIABILITY_THRESHOLD,
                    )

                base = str(data_storage / aquifer / frame / 'mintpy')

                with h5py.File(f'{base}/geometryGeo.h5', 'r') as geom:
                    epsg = dict(geom.attrs)['EPSG']
                transformer = Transformer.from_crs(f"EPSG:{epsg}", "EPSG:4326", always_xy=True)

                with xr.open_dataset(f'{base}/avgSpatialCoh.h5', engine='h5netcdf') as coh_ds:
                    coh_ds = coh_ds.rename({'phony_dim_0': 'y', 'phony_dim_1': 'x'})
                    coherence = coh_ds.avgSpatialCoh

                    stacked = coherence.stack(z=("y", "x"))
                    max_flat_idx = int(stacked.argmax().values)
                    y_idx, x_idx = np.unravel_index(max_flat_idx, coherence.shape)

                    y_coord = coh_ds["y"].isel(y=y_idx).values.item()
                    x_coord = coh_ds["x"].isel(x=x_idx).values.item()

                ref_lon, ref_lat = transformer.transform(x_coord, y_coord)
                ref_lat = float(ref_lat)
                ref_lon = float(ref_lon)
                ref_y = int(y_idx)
                ref_x = int(x_idx)

                attrs = {
                    'REF_LAT': ref_lat,
                    'REF_LON': ref_lon,
                    'REF_Y':   ref_y,
                    'REF_X':   ref_x,
                }

                for fname in ['velocity.h5', 'timeseries.h5', 'geometryGeo.h5']:
                    fpath = f'{base}/{fname}'
                    with h5py.File(fpath, 'r+') as f:
                        f.attrs.update(attrs)
                    print(f"Added to {fpath}")
                    print(f"  REF_LAT: {ref_lat:.6f}, REF_LON: {ref_lon:.6f}, REF_Y: {ref_y}, REF_X: {ref_x}")

                shutil.rmtree(data_storage / aquifer / frame / 'geom_data')
                shutil.rmtree(data_storage / aquifer / frame / 'orbit_data')
                shutil.rmtree(data_storage / aquifer / frame / 'subset-ncs')
```

### We need to assign a reference location based on coherence so we can decompose 
- This is the location with the highest average temporal coherence

```python
if aquifer:
    for frame in os.listdir(data_storage / aquifer):
        if frame == '3067':
            continue
        else:
            base = str(data_storage / aquifer / frame / 'mintpy')

            with h5py.File(f'{base}/geometryGeo.h5', 'r') as geom:
                epsg = dict(geom.attrs)['EPSG']
            transformer = Transformer.from_crs(f"EPSG:{epsg}", "EPSG:4326", always_xy=True)

            with xr.open_dataset(f'{base}/avgSpatialCoh.h5', engine='h5netcdf') as coh_ds:
                coh_ds = coh_ds.rename({'phony_dim_0': 'y', 'phony_dim_1': 'x'})
                coherence = coh_ds.avgSpatialCoh

                stacked = coherence.stack(z=("y", "x"))
                max_flat_idx = int(stacked.argmax().values)
                y_idx, x_idx = np.unravel_index(max_flat_idx, coherence.shape)

                y_coord = coh_ds["y"].isel(y=y_idx).values.item()
                x_coord = coh_ds["x"].isel(x=x_idx).values.item()

            ref_lon, ref_lat = transformer.transform(x_coord, y_coord)
            ref_lat = float(ref_lat)
            ref_lon = float(ref_lon)
            ref_y = int(y_idx)
            ref_x = int(x_idx)

            attrs = {
                'REF_LAT': ref_lat,
                'REF_LON': ref_lon,
                'REF_Y':   ref_y,
                'REF_X':   ref_x,
            }

            for fname in ['velocity.h5', 'timeseries.h5', 'geometryGeo.h5']:
                fpath = f'{base}/{fname}'
                with h5py.File(fpath, 'r+') as f:
                    f.attrs.update(attrs)
                print(f"Added to {fpath}")
                print(f"  REF_LAT: {ref_lat:.6f}, REF_LON: {ref_lon:.6f}, REF_Y: {ref_y}, REF_X: {ref_x}")

else:
    for aquifer in os.listdir(data_storage):
        for frame in os.listdir(data_storage / aquifer):
            base = str(data_storage / aquifer / frame / 'mintpy')
        
            with h5py.File(f'{base}/geometryGeo.h5', 'r') as geom:
                epsg = dict(geom.attrs)['EPSG']
            transformer = Transformer.from_crs(f"EPSG:{epsg}", "EPSG:4326", always_xy=True)
        
            with xr.open_dataset(f'{base}/avgSpatialCoh.h5', engine='h5netcdf') as coh_ds:
                coh_ds = coh_ds.rename({'phony_dim_0': 'y', 'phony_dim_1': 'x'})
                coherence = coh_ds.avgSpatialCoh
        
                stacked = coherence.stack(z=("y", "x"))
                max_flat_idx = int(stacked.argmax().values)
                y_idx, x_idx = np.unravel_index(max_flat_idx, coherence.shape)
        
                y_coord = coh_ds["y"].isel(y=y_idx).values.item()
                x_coord = coh_ds["x"].isel(x=x_idx).values.item()
        
            ref_lon, ref_lat = transformer.transform(x_coord, y_coord)
            ref_lat = float(ref_lat)
            ref_lon = float(ref_lon)
            ref_y = int(y_idx)
            ref_x = int(x_idx)
        
            attrs = {
                'REF_LAT': ref_lat,
                'REF_LON': ref_lon,
                'REF_Y':   ref_y,
                'REF_X':   ref_x,
            }
        
            for fname in ['velocity.h5', 'timeseries.h5', 'geometryGeo.h5']:
                fpath = f'{base}/{fname}'
                with h5py.File(fpath, 'r+') as f:
                    f.attrs.update(attrs)
                print(f"Added to {fpath}")
                print(f"  REF_LAT: {ref_lat:.6f}, REF_LON: {ref_lon:.6f}, REF_Y: {ref_y}, REF_X: {ref_x}")
```
